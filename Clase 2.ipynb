## Resolver la ecuación diferencial

$X''-λX=0 \quad X'(0)=X'(1)=0$

Importamos la librerias necesarias

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

 Definir variables simbólicas

```python
x = sp.symbols('x')
n = sp.symbols('n', integer=True, nonnegative=True)
```

Eigenvalores y eigenfunciones

```python
lambda_n = -(n * sp.pi)**2
X_n = sp.cos(n * sp.pi * x)
```

Configurar gráficas

```python
plt.figure(figsize=(10, 6))
x_vals = np.linspace(0, 1, 1000)
```

```python
print("Soluciones para cada n:")
print("-----------------------")
```

Graficar para n = 0, 1, 2, 3

```python
for n_val in range(4):
    # Sustituir n por el valor actual
    X_n_eval = X_n.subs(n, n_val)
    lambda_n_eval = lambda_n.subs(n, n_val)
    # Imprimir soluciones en formato simbólico
    print(f"\nn = {n_val}:")
    sp.pprint(sp.Eq(sp.Symbol("\lambda"), lambda_n_eval, evaluate=False))  # λ sin evaluar
    sp.pprint(sp.Eq(sp.Symbol(f"X_{n_val}(x)"), X_n_eval, evaluate=False))  # X_n(x) sin evaluar

    # Convertir a función numérica
    X_func = sp.lambdify(x, X_n_eval, 'numpy')
    y_vals = X_func(x_vals)

    # Corregir el caso n=0 (evitar escalar)
    if n_val == 0:
        y_vals = np.full_like(x_vals, y_vals)

    plt.plot(x_vals, y_vals, label=rf'$n={n_val}$, $\lambda={lambda_n_eval.evalf():.2f}$')

plt.title('Soluciones de $X\'\' - \lambda X = 0$ con $X\'(0)=X\'(1)=0$')
plt.xlabel('x')
plt.ylabel('$X_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

Resolver

$X''-λX=0 \quad X(0)=X(1)=0$
