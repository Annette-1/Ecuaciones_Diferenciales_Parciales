```python
#Ejercicio graficar y=x²
import matplotlib.pyplot as plt
import numpy as np

# Definir el rango de valores para x
x = np.linspace(-10, 10, 400)

# Calcular los valores de y = x^2
y = x**2

# Crear la gráfica
plt.plot(x, y, label='y = x²')

# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de y = x²')

# Añadir una cuadrícula
plt.grid(True)

# Mostrar la leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
