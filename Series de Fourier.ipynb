\end{cases}
  
La solucion de la serie de Fourier para f(x) esta dada por:
$$u(x,t)= \frac{1}{2} + \sum_{n=1}^{\infty} \left( \frac{2}{(n\pi)^2} (1 - (-1)^n) \cos \left( \frac{n\pi x}{2} \right) +  \frac{2}{n\pi} sen ( \frac{n\pi x}{2} \right).$$
