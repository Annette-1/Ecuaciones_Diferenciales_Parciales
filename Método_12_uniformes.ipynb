```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def gaussiana (mu,Sigma,n):
   muestra=[]
   for i in range (n):
     Suma= 0
     for j in range(12):
        u=r.random()
        Suma=Suma+u
     x= mu +Sigma*(Suma-6)
     muestra.append(x)
   return muestra
```

```python
M=gaussiana(0,1,100000)
```

```python
plt.hist(M,bins=50)
```

# **Método de Box-Muller**

```python
def muller(n):
  muestra2=[]
  for i in range (n):
     u1=r.random()
     u2=r.random()
     X=np.sqrt(-2* np.log(u1)) * np.cos(2*np.pi*u2)
     Y=np.sqrt(-2* np.log(u1)) * np.sin(2*np.pi*u2)
     muestra2.append(X)
     muestra2.append(Y)
  return muestra2
```

```python
L=muller(100000)
```

```python
plt.hist(L,bins=50)
```
