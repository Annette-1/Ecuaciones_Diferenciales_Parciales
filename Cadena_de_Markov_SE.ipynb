# <span style="color:purple;"> Cadena de Markov</span>

Algoritmo para simular una CM sobre S:

Entrada: $α,P,n$

Salida :$X_0,X_1,...,X_n$

Generar: $X_0 ∼α$

Para $i=1,...,n$

Bucle:
si $$X_{i-1}=j$$
tomar $$p=\quad renglón \quad j \quad de \quad P$$
generar
$$X_i ∼P$$

Estados $S={0,1,2.3}$
Salida una sucesión de estados  ${0,1,1,3,2,1,...}$

\end{bmatrix}
$$

```python
import numpy as np
import random
```

Vamos a crear una funcion para simular una cadena de Markov con los parametros


Parámetros:
* alpha: distribución inicial (lista de probabilidades)

* P: matriz de transición (lista de listas)

* n: número de pasos a simular
    
* Retorna:

Lista con la trayectoria $X_0, X_1, ..., X_n$ (Sucesión)


```python
def simular_cadena_markov(alpha, P, n):
    """
    Simula una cadena de Markov usando random.choices

    Parámetros:
    alpha: distribución inicial (lista de probabilidades)
    P: matriz de transición (lista de listas)
    n: número de pasos a simular

    Retorna:
    Lista con la trayectoria X0, X1, ..., Xn (estados 0-based)
    """

    # Número de estados
    num_estados = len(alpha)
    estados = list(range(num_estados))

    # Generar X0 según distribución alpha
    X0 = random.choices(estados, weights=alpha, k=1)[0]
    X = [X0]

    # Simular los siguientes pasos
    for i in range(n):
        estado_actual = X[-1]

        # Obtener la fila de transición correspondiente
        fila_transicion = P[estado_actual]

        # Generar siguiente estado usando random.choices
        siguiente = random.choices(estados, weights=fila_transicion, k=1)[0]
        X.append(siguiente)

    return X
```

Retomamos los datos del ejercicio

```python
alpha = [1/4, 1/4, 1/4, 1/4] # Distribución inicial uniforme
P = [
    [1/4, 0, 1/4, 2/4],
    [2/4, 1/4, 0, 1/4],
    [1/4, 2/4, 1/4, 0],
    [0, 1/4, 2/4, 1/4]
]
n = 25
```

```python
# Fijar semilla para reproducibilidad (es opcional se puede omitir)
random.seed(42)
```

```python
# Ejecutar simulación
trayectoria = simular_cadena_markov(alpha, P, n)
```

```python
# Mostrar resultado como lista
print(" Sucesión de S :")
print(trayectoria)
```

```python
# Contamos la probabilidad para cada estado
probabilidades = [trayectoria.count(i) / len(trayectoria) for i in range(len(P))]
```

```python
import numpy as np

def matriz_transicion_serpientes_escaleras():
    # Configuración
    n = 20
    dado = 6

    # Saltos especiales
    salto = list(range(n + 1))  # índice 0 no se usa
    # Escaleras
    salto[3] = 11
    salto[15] = 19
    # Serpientes
    salto[13] = 5
    salto[17] = 7

    # Crear matriz
    P = np.zeros((n, n))

    for i in range(1, n + 1):
        if i == n:  # Casilla 20 absorbente
            P[i-1, i-1] = 1
            continue

        for k in range(1, dado + 1):
            # Posición después del dado
            pos = i + k
            if pos > n:
                pos = i  # Se queda si se pasa
            # Aplicar salto si existe
            pos_final = salto[pos]
            P[i-1, pos_final-1] += 1/dado

    return P

# Obtener la matriz
P = matriz_transicion_serpientes_escaleras()
print("Matriz de transición:")
print(P)
```

```python
import pandas as pd

P = matriz_transicion_serpientes_escaleras()
df = pd.DataFrame(P,
                  index=[f"Desde {i+1}" for i in range(20)],
                  columns=[f"Hasta {j+1}" for j in range(20)])

# Mostrar solo algunas filas de ejemplo
print("Ejemplos de filas:")
print(df.iloc[12:18, :])  # Filas 13-18
```
