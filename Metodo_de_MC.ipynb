```python
def test_function(x):
  return np.sin(x)
```

```python
def integrate(x1,x2,func=test_function,n=100000):
    X=np.linspace(x1,x2,1000)
    y1=0
    y2=max((func(X)))
    print(x1,x2,y1,y2)
    area=(x2-x1)*(y2-y1)
    check=[]
    xs=[]
    ys=[]
    for i in range(n):
        x=np.random.uniform(x1,x2,1)
        xs.append(x)
        y=np.random.uniform(y1,y2,1)
        ys.append(y)
        if abs(y)>abs(func(x)) or y<0:
            check.append(0)
        else:
            check.append(1)
    return(np.mean(check)*area,xs,ys,check)
```

```python
#print(integrate(0,1)[0])
print("Valor estimado de la integral:",integrate(0,1,test_function)[0])
_,x,y,c=integrate(0,1,n=500)
df=pd.DataFrame()
df['x']=x
df['y']=y
df['c']=c

X=np.linspace(0,1,1000)
plt.grid(True)
plt.plot(X,test_function(X))
plt.scatter(df[df['c']==0]['x'],df[df['c']==0]['y'],color='green')
plt.scatter(df[df['c']==1]['x'],df[df['c']==1]['y'],color='blue')
plt.show()
```
